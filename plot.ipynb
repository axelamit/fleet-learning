```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("history.pkl", "rb") as f:
    hist = pickle.load(f)

loss = [l[1] for l in hist.losses_centralized]
x = np.arange(len(loss))
plt.plot(x, loss)
plt.title("Fed loss")
plt.xlabel("Global round")
plt.ylabel("Loss")
plt.show()
```
