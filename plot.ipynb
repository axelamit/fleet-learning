```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
#date = "27-06-2023-11:38:51"
date = "27-06-2023-14:38:16"

with open(f"results/{date}/history.pkl", "rb") as f:
    hist = pickle.load(f)

loss = [l[1] for l in hist.losses_centralized]
x = np.arange(len(loss))
plt.plot(x, loss)
plt.title("Fed loss")
plt.xlabel("Global round")
plt.ylabel("Loss")
plt.show()
print(loss)
```
